```python
import requests
import json
import os

os.makedirs("data/raw", exist_ok=True)

print("Pronto!")
```

```python
def buscar_deputados():

    Todos_Deputados = []
    url = "https://dadosabertos.camara.leg.br/api/v2/deputados"

    try:

        resposta = requests.get(url, timeout=10)
        dados = resposta.json()
        Todos_Deputados.extend(dados['dados'])
        print("Carga de deputados executada com sucesso")

    except Exception as e:
        print("Carga de deuptados não carregada por erro na api")

    return Todos_Deputados
```

```python
deputados = buscar_deputados()
```

```python
with open ("data/raw/deputados.json", "w", encoding="utf-8") as f:
    json.dump(deputados, f, indent=2)
print(f"Total de {len(deputados)} carregados")
```
