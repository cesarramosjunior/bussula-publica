```python
import requests
import json
import os
import time


# Pasta onde vou armezenar os arquivos json
os.makedirs("data/raw", exist_ok=True)

print ("Pasta criada com sucesso!")
```

```python
def buscar_partidos ():

    todos_partidos = []
    pagina = 1

    while True:

        url = f"https://dadosabertos.camara.leg.br/api/v2/partidos?pagina={pagina}&itens=15"

        try:

            resposta = requests.get(url, timeout=10)
            dados = resposta.json()
            if not (dados['dados']):
                break

            todos_partidos.extend(dados['dados'])
            print (f"Página {pagina} com {len(dados['dados'])} partidos carregada com sucesso!")

            pagina += 1
            time.sleep(1)


        except Exception as e:
            print (f"Erro na página: {pagina}: {e}")  
            break

    return todos_partidos

                
```

```python
partidos = buscar_partidos ()
```

```python
with open ("data/raw/partidos.json", "w", encoding="utf-8") as f:
    json.dump(partidos, f, ensure_ascii=False, indent=2)

print(f"Salvo total de {len(partidos)} partidos.")
```
