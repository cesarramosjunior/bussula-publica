```python
import requests
import json
import os
import time

# Pasta onde vamos salvar os JSONs brutos
os.makedirs("data/raw", exist_ok=True)

print("Pasta criada com sucesso!")
```

```python
def buscar_proposicoes(data_inicio, data_fim):
    todas_proposicoes = []
    pagina = 1

    while True:
        url =f"https://dadosabertos.camara.leg.br/api/v2/proposicoes?dataInicio={data_inicio}&dataFim={data_fim}&pagina={pagina}&itens=100"

        try:
            resposta = requests.get(url, timeout=10)
            dados = resposta.json()

            if not dados['dados']:
                break

            todas_proposicoes.extend(dados['dados'])
            print(f"Pagina {pagina} coletada - {len(todas_proposicoes)} proposições até agora")

            pagina += 1
            time.sleep(1) # espera meio segundo entre páginas

        except Exception as e:
            print(f"Erro na página {pagina}: {e}")
            break

    return todas_proposicoes

```

```python
proposicoes = buscar_proposicoes("2025-05-15", "2025-05-22")
```

```python
with open("data/raw/proposicoes.json", "w", encoding="utf-8") as f:
    json.dump(proposicoes, f, ensure_ascii=False, indent=2)

print(f"Salvo! Total de {len(proposicoes)} proposições.")
```
