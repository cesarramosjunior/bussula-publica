```python
import requests
import json
import os
import time

os.makedirs("data/raw", exist_ok=True)

print("Pronto!")
```

```python
with open("data/raw/deputados.json", "r", encoding="utf-8") as f:
    deputados = json.load(f)

print(f"Total de {len(deputados)} deputados carregados!")
```

```python
def busca_despesas (ano, mes):

    todas_despesas = []
    
    for deputado in deputados[:10]:

        id = deputado['id']
        pagina = 1

        while True:

            url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{id}/despesas?ano={ano}&mes={mes}&pagina={pagina}&itens=100"

            try:

                resposta = requests.get(url, timeout=30)
                dados = resposta.json()
                if not dados['dados']:
                    break

                for despesa in dados['dados']:
                    despesa['id_deputado'] = id

                todas_despesas.extend(dados['dados'])
                print(f"Deputado {id} - Despesas carregadas - Página: {pagina}")

                pagina += 1
                time.sleep(3) 

            except Exception as e:
                print (f"Erro para o deputado {id} na página {pagina}: {e}")
                time.sleep(5)
                continue


    return todas_despesas
```
