```python
import requests
import json
import os
import time

# Pasta onde vamos salvar os JSONs brutos
os.makedirs("data/raw", exist_ok=True)

print("Pasta criada com sucesso!")
```

```python
def busca_votacoes (data_inicio, data_fim):
    todas_votacoes = []
    pagina = 1

    while True:

        url = f"https://dadosabertos.camara.leg.br/api/v2/votacoes?dataInicio={data_inicio}&dataFim={data_fim}&pagina={pagina}&itens=100"

        try:

            resposta = requests.get(url, timeout = 30)
            dados = resposta.json()
            if not dados['dados']:
                break

            todas_votacoes.extend(dados['dados'])
            print(f"Pagina {pagina} coletada - total de {len(todas_votacoes)} votações até agora")

            pagina += 1
            time.sleep(3) # espera meio segundo entre páginas


        except Exception as e:
            print(f"Erro na página {pagina}: {e}")
            time.sleep(5)
            continue

    return todas_votacoes

```

```python
votacoes = busca_votacoes ("2025-05-15", "2025-05-22")
```

```python
with open("data/raw/votacoes.json", "w", encoding="utf-8") as f:
    json.dump(votacoes, f, ensure_ascii=False, indent=2)

print(f"Salvo! Total de {len(votacoes)} votações.")
```
