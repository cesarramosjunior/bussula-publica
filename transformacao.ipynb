```python
import pandas as pd
import json
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

print("Pronto!")
```

```python
load_dotenv()
DATABASE_URL = os.getenv("DATABASE_URL")
engine = create_engine(DATABASE_URL)

print("Conexão criada com sucesso!")
```

```python
with open("data/raw/deputados.json", "r", encoding="utf-8") as f:
    deputados = json.load(f)

df_deputados = pd.DataFrame(deputados)[['id', 'nome', 'siglaPartido', 'siglaUf', 'idLegislatura', 'email']]


df_deputados.head()
```

```python
df_deputados.to_sql('deputados', engine, if_exists='replace', index=False)

print("Deputados carregados no banco!")
```
